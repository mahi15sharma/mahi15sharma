```python
import seaborn as sns #  Python data visualization library based on matplotlib for statistical graphics
import pandas as pd # data analysis and manipulation tool
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fmri=sns.load_dataset('fmri')
fmri.head(5)
sns.relplot(x = 'timepoint', y = 'signal', data = fmri)
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, sort = True)
sns.catplot(x = 'timepoint', y = 'signal', kind = 'box', data = fmri)
```
